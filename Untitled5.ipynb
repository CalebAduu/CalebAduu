```python
df = pd.read_csv('indexData.csv', encoding = 'latin-1')
```

```python
df.head()
```

```python
grouped = df.groupby('Element')
for element, group in grouped:
    print("Element:", element)
    print(group)
    print()
```

```python
# Filter the dataset for the year 2017
df_2017 = df[df['Y2017'] == 2017]

# Convert numeric columns to float and handle non-numeric values
df_2017_numeric = df_2017.apply(pd.to_numeric, errors='coerce')

# Remove rows with non-numeric values
df_2017_numeric = df_2017_numeric.dropna()

# Calculate the mean and standard deviation for the filtered dataset
mean_2017 = df_2017_numeric.mean().round(2)
std_2017 = df_2017_numeric.std().round(2)

# Display the mean and standard deviation
print("Mean (2017):\n", mean_2017)
print("\nStandard Deviation (2017):\n", std_2017)
```

```python
selected_columns = df[['Y2017', 'Area']]

# Perform a groupby operation on 'Area'
grouped_data = selected_columns.groupby('Area')

# Display the grouped data
for area, data in grouped_data:
    print("Area:", area)
    print(data)
    print("\n")
    
```

```python
df_2017 = df[df['Y2017'].notnull()]

# Calculate the sum of 'Y2017' for each area
area_sum_2017 = df_2017.groupby('Area')['Y2017'].sum()

# Find the area with the lowest sum in 2017
area_with_lowest_sum = area_sum_2017.idxmin()
lowest_sum = area_sum_2017.min()

# Display the area with the lowest sum and its corresponding sum value
print("Area with the lowest sum in 2017:", area_with_lowest_sum)
print("Lowest sum in 2017:", lowest_sum)
```

```python
df_ghana_2017 = df[(df['Area'] == 'Ghana') & (df['Y2017'].notnull())]

# Calculate the sum of 'Y2017' for the specific area 'Ghana'
sum_ghana_2017 = df_ghana_2017['Y2017'].sum()

# Display the total sum for the specific area 'Ghana'
print("Total sum for Ghana in 2017:", sum_ghana_2017)
```
